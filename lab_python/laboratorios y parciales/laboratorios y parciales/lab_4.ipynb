```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("All_Pokemon.csv", index_col='Number')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
duplicados = df.duplicated().sum()
if duplicados > 0:
    print(f"Se encontraron duplicados:{duplicados},procediendo a eliminarlos")
    df.drop_duplicates()
else:
    print("No se encontraro duplicados")
print("Dimensiones tras el paso de los duplicados:",df.shape)
```

```python
nulos_columnas =df.isnull().sum()
nulos_filas = df[df.isnull().any(axis=1)]
total_nulos = nulos_columnas.sum()
print(total_nulos)
```

```python
nulos_columnas
```

```python
if nulos_columnas['Type 2'] > 0:
    df.dropna(subset=["Type 2"], inplace=True)
    print("Limpiesa de nulos completada")
else:
    print("No hay valores nulos que limpiar")
```

```python
df.info()
```

```python
def filtro_fire(type): # funcion que filtra los fuego
    return 'Type 1' in type
```

```python
# aplicar un filtro para consultar y llamar la funciÃ³n filtro_fire
datos_fuego= df[df['Type 1'].apply(filtro_fire)]
datos_fuego.head(20)
```
