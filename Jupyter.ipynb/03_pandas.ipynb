**crear una funci√≥n para verificar si un video juego esta presente en la lista**

en el ejemplo vamos a verificar si se encuentra el genero RPG en la lista

```python
def filtro_rpg(genres): # funcion que filtra los rpg
    return 'RPG' in genres
```

```python
# aplicar un filtro para consultar y llamar la funci√≥n filtro_rpg
datosRPG = datos[datos['Genres'].apply(filtro_rpg)]
```

```python
# mostrar el dataset filtardo; incluyendo solo las 2 columnas titulo y rating
nombreRPG = datosRPG[['Title', 'Rating' ]]
nombreRPG.head(20) # muestra los primeros 20 juegos
```

```python
nombreRPG.info()
```

**Graficar los primeros 10 video juego y su rating de popularidad**

```python
graficosRPG = nombreRPG.head(10) #grafico solo tiene los 10 primeros video juego
graficosRPG.drop(['Title'],axis=1)['Rating'].plot(kind='bar')
```

**Mejorando el gr√°fico**

```python
graficosRPG = nombreRPG.head(10) #grafico solo tiene los 10 primeros video juego
graficosRPG.set_index('Title', inplace=True) # establece la colunma titulo como indice
graficosRPG['Rating'].plot(kind='bar')
```

**Agregando etiquetas al gr√°fico**

```python
graficosRPG = nombreRPG.head(10) #grafico solo tiene los 10 primeros video juego
graficosRPG.set_index('Title', inplace=True) # establece la colunma titulo como indice
graph = graficosRPG['Rating'].plot(kind='bar')
# agregar las etiquetas n√∫mericas de rating en las barras
for i, rating in enumerate(graficosRPG['Rating']):
    graph.annotate(str(rating), xy=(i,rating), ha='center')
```

**Gr√°fico de Torta**

**Promedio  de Rating de video Juegos RPG**

```python
promedioRPG = datosRPG[['Rating']] #promedio de video juego PRG
promRPG = promedioRPG.mean() #funci√≥n promedio numpy
```

**Promedio de Rating de video juego Indie**

```python
def filtro_indie(genres):  # funci√≥n que filtra el genero indie
    return 'Indie' in genres
```

```python
#aplicar un filtro para consultar y llamar  la funcion filtro_indie
datosIndie = datos[datos['Genres'].apply(filtro_indie)] #creando un nuevo dataframe con los viedo juegos Indie
```

```python
# Mostrando el datafreme filtrado incluyendo solo las columna (Title y Rating)
nombresIndie = datosIndie[['Title','Rating']]
nombresIndie.head(20)
```

```python
promedioIndie = datosIndie[['Rating']] #promedio del video juego RPG
promIND =promedioIndie.mean()  #saque el promedio y lo guarde en una variable
```

###**Crear El Gr√°fico Circular**

```python
dataset_promedio = pd.DataFrame({'Genero':['RPG','Indie'],'Promedio':['promRPG','promIND']})
```
