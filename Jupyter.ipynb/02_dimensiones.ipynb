**Dos vectores en la misma celda**

```python
vector1 = np.random.randint(1,11,12) # agregrar el grafico y grafico de barras
vector2 = np.random.randint(5,21,12)
```

```python


plt.plot( vector1, marker = 'o' ,label = "Vector 1")
plt.plot(vector2,label = "Vector 2")
plt.title("Comparando dos Vectores")
plt.xlabel("Indice (Posicin)")
plt.ylabel("Valores")
plt.grid(True)
plt.legend("vector")
plt.show
```

**Creando Grafico de Barras**

```python
plt.bar(range(len(vector2)),vector1)
```
